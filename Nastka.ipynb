```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("./real_fake/Data_short.csv", parse_dates=['date'])
```

```python
subject_counts = data['subject'].value_counts()
plt.bar(subject_counts.index, subject_counts)
plt.xticks(rotation='vertical')
plt.xlabel('Subject')
plt.ylabel('Count')
plt.title('Histogram of Subjects')
plt.show()
```

```python
df = pd.read_csv('real_fake/Data_short.csv')
df['date'] = pd.to_datetime(df['date'])
df['day_of_week'] = df['date'].dt.day_name()
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df['day_of_week'] = pd.Categorical(df['day_of_week'], categories=days_order, ordered=True)
result = df.groupby(['day_of_week', 'is_true']).size().unstack().fillna(0)
result.plot(kind='bar', stacked=True)
plt.title('Count of Real and Fake News for Each Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Count')
plt.legend(['Fake', 'Real'])
plt.show()
```
