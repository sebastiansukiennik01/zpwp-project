```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("./real_fake/Data_short.csv", parse_dates=['date'])
```

```python
data.show()
```

```python
subject_counts = data['subject'].value_counts()
plt.bar(subject_counts.index, subject_counts)
plt.xticks(rotation='vertical')
plt.xlabel('Subject')
plt.ylabel('Count')
plt.title('Histogram of Subjects')
plt.show()
```
