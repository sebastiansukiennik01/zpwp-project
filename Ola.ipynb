```python
df = pd.read_csv('real_fake\Data_short.csv')
df['date'] = pd.to_datetime(df['date'])
df['month'] = df['date'].dt.to_period('M')
result = df.groupby(['month', 'is_true']).size().unstack().fillna(0)

result.plot(kind='bar', stacked=True)
plt.title('Count of Real and Fake News for Each Month')
plt.xlabel('Month')
plt.ylabel('Count')
plt.legend(['Fake', 'Real'])
plt.show()
```

```python
df = df[df['date'].dt.year != 2015]
df['month'] = df['date'].dt.to_period('M')
result = df.groupby(['month', 'is_true']).size().unstack().fillna(0)

# Plot the data
result.plot(kind='bar', stacked=True)
plt.title('Count of Real and Fake News for Each Month (excluding 2015)')
plt.xlabel('Month')
plt.ylabel('Count')
plt.legend(['Fake', 'Real'])
plt.show()
```

```python
subject_counts = df.groupby(['subject', 'is_true']).size().unstack(fill_value=0)

# Plot the data
subject_counts.plot(kind='bar', stacked=True)
plt.title('Count of True and False News for Each Subject')
plt.xlabel('Subject')
plt.ylabel('Count')
plt.legend(['False', 'True'])
plt.show()
```
